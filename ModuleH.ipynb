# **22. Ordinary Differential Equation - Initial Value Problems**

What are **differential equations**?
They are relationships between a funtion and its derivatives. They can be used to model systems in every STEM field. We know that finding exact solutions to these equations can be difficult whereas for differentiation, analytical solutions can be computed. Therefore, numerical solutions are crucial to make these equations useful for designing and understanding systems.

## **22.1 ODE Initial Value Problem Statement**

A **differential equation** is a relationship between a function, f(x), its independent variable, x, and any number of its derivatives. Differential equations can be categorised into two types: ordinary and partial equations.

An **ordinary differential equation** or ODE is a differential equation where the independent variable, and the derivatives, are in one dimension. For an example of an ODE, consider a pendulum of length ùëô with a mass, ùëö, at its end. So, $ml\frac{d^2\Theta(t)}{dt^2} = -mg \sin(\Theta(t))$, is the angle the pendulum makes with the vertical axis over time, Œò(ùë°), in the presence of vertical gravity, ùëî.

A **partial differentiation equation** or PDE is  a general form differential equation where ùë• is a vector containing the independent variables $x_1, x_2, x_3, \ldots, x_m$, and the partial derivatives can be of any order and with respect to any combination of variables. An example of a PDE would be the heat equation which describes the evolution of temperature in space over time. So, $\frac{\partial u(t, x, y, z)}{\partial t} = \alpha\left(\frac{\partial u(t, x, y, z)}{\partial x} + \frac{\partial u(t, x, y, z)}{\partial y} + \frac{\partial u(t, x, y, z)}{\partial z}\right)$, where ùë¢(ùë°,ùë•,ùë¶,ùëß) is the  (ùë•,ùë¶,ùëß) at time ùë°, and ùõº is a thermal diffusion constant.$

An equation is referred to as the $n^{\mathrm{th}}$ order ODE where n is the **order** of differential equation. For example, a differential equation can be written as $F\left(x, f(x), \frac{df(x)}{dx}, \frac{d^2f(x)}{dx^2}, \frac{d^3f(x)}{dx^3}, \ldots, \frac{d^{n-1}f(x)}{dx^{n-1}}\right) = \frac{d^{n}f(x)}{dx^{n}}.$

A **general solution** to a differential equation is a ùëî(ùë•) that satisfies the differential equation. For an ODE of order ùëõ, a **particular solution** is a ùëù(ùë•) that satisfies the differential equation and n explicitly **known values** of the solution, or its derivatives, at certain points.

A common set of known values for an ODE solution is the **initial value**. Finding a solution to an ODE given an initial value is called the **initial value problem**.

## **22.2 Reduction of Order**

There are numerous numerical methods available to solve initial value problems, however, there is a catch. They are designed to particularly solve first-order differential equations. In order to solve higher order differential equations, the order of the differential equation in question must be reduced to first order. When we reduce the order of an ODE to first order, it can result in multiple variables. But since the derivatives are still taken with respect to the same independent variable, the ordinariness of the differential equation is retained.


***A simple model to describe the change in population of rabbits, ùëü(ùë°), and wolves, ùë§(ùë°), might be $\frac{dr(t)}{dt} = 4r(t) - 2w(t)  and $$\frac{dw(t)}{dt} = r(t) + w(t)$. Write this system of differential equations as an equivalent differential equation in ùëÜ(ùë°).***

$\begin{split}
S(t) =\left[\begin{array}{c}
r(t) \\
w(t)
\end{array}\right].
\end{split}$

This first order ODE is equivalent to the pair of ODEs.
$\begin{split}
\frac{dS(t)}{dt} = \left[\begin{array}{cc}
4 & -2 \\
1 & 1
\end{array}\right]S(t).
\end{split}$

## **22.3 The Euler Method**

The **Explicit Euler Formula** is $S(t_{j+1}) = S(t_j) + hF(t_j, S(t_j))$. It allows us to compute an approximation for the state at $S(t_{j+1})$ given the state at $S(t_j)$.   

***The differential equation $\frac{df(t)}{dt} = e^{-t}$ with initial condition $ùëì_0 = ‚àí1$ has the exact solution $f(t) = -e^{-t}$. Approximate the solution to this initial value problem between 0 and 1 in increments of 0.1 using the Explicity Euler Formula. Plot the difference between the approximated solution and the exact solution.***

```python
import numpy as np
import matplotlib.pyplot as plt

# Defining parameters.
f = lambda t, s: np.exp(-t) # This is the ODE.
h = 0.1 # Step size.
t = np.arange(0, 1 + h, h) # Numerical grid.
s0 = -1 # Initial Condition.

# Explicit Euler Method.
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (4, 4))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

Repeating the above process for $h = 0.01$, we can get a better approximation of the solution on the numerical grid.. we can see each dot is one approximation based on the previous dot in a linear fashion.

```python
h = 0.01 # Step size.
t = np.arange(0, 1 + h, h) # Numerical grid.
s0 = -1 # Initial Condition.

# Explicit Euler Method.
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (4, 4))
plt.plot(t, s, 'b--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
